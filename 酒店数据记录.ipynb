```python
# 归一化
def byone(x):
    max = np.max(x)
    min = np.min(x)
    x1 = (x-min)/(max-min)
    return x1
# 对hotel进行归一化处理
a = hotel['base_price']
b = hotel['hotel_latitude']
c = hotel['hotel_longitude']

hotel['base_price'] = byone(a)
hotel['hotel_latitude'] = byone(b)
hotel['hotel_longitude'] = byone(c)

hotel
```

```python
def byone(x):
    max = np.max(x)
    min = np.min(x)
    x1 = (x-min)/(max-min)
    return x1
# 对customer进行归一化处理
a = customer['age']
b = customer['sex']
c = customer['home_latitude']
d = customer['home_longitude']
customer['age'] = byone(a)
customer['sex'] = byone(b)
customer['home_latitude'] = byone(c)
customer['home_longitude'] = byone(d)
customer
```

```python
# 对customer进行降维处理，变成二维
# 删除非数字的指标
customer = customer.drop('customer_id',axis = 1)
```

```python
# 对customer进行降维处理，变成二维
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
customer = customer.dropna()
y = pca.fit_transform(customer)
y
```
