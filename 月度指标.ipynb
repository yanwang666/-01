```python
monthly_index = monthly_index.drop('year_month',axis = 1)
monthly_index
```

```python
#进行归一化0
```

```python
# 归一化
def byone(x):
    max = np.max(x)
    min = np.min(x)
    x1 = (x-min)/(max-min)
    return x1
# 进行归一化处理
a = monthly_index['sales_amount']
b = monthly_index['customer_number']
monthly_index['sales_amount'] = byone(a)
monthly_index['customer_number'] = byone(b)
monthly_index
```

```python
# 对monthly_index进行降维处理，变成二维
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
monthly_index = monthly_index.dropna()
y = pca.fit_transform(monthly_index)
y
```

```python
#实现数据的可视化
pit.scatter(y[:,0],y[:,1],marker = '*')
pit.show()
```
