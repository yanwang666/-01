```python
# 归一化
def byone(x):
    max = np.max(x)
    min = np.min(x)
    x1 = (x-min)/(max-min)
    return x1
# 进行归一化处理
a = production['length']
b = production['thickness']
production['length'] = byone(a)
production['thickness'] = byone(b)
production
```

```python
# 对production进行降维处理，变成二维
# 删除非数字的指标
production = production.drop(['type','fault_flg'],axis = 1)
```

```python
# 对customer进行降维处理，变成二维
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
production = production.dropna()
y = pca.fit_transform(production)
y
```

```python
#实现数据的可视化
pit.scatter(y[:,0],y[:,1],marker = '*')
pit.show()
```
